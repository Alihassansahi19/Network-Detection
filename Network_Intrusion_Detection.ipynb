```python
X = df.loc[:,['Received Packets','Received Bytes','Sent Bytes','Sent Packets','Port alive Duration (S)',
              'Packets Rx Dropped','Packets Tx Dropped','Packets Rx Errors','Packets Tx Errors',
              'Delta Received Packets','Delta Received Bytes','Delta Sent Bytes','Delta Sent Packets',
              'Delta Port alive Duration (S)', 'Delta Packets Rx Dropped',
       ' Delta Packets Tx Dropped', 'Delta Packets Rx Errors',
       'Delta Packets Tx Errors', 'Connection Point', 'Total Load/Rate',
       'Total Load/Latest', 'Unknown Load/Rate', 'Unknown Load/Latest',
       'Latest bytes counter','Table ID', 'Active Flow Entries',
       'Packets Looked Up', 'Packets Matched', 'Max Size',]]
```

```python
#Label Encoding of Values( Example : Male = 0  and Female = 1)
y = df.iloc[:,-2].values
```

```python
from sklearn.preprocessing import LabelEncoder
ly = LabelEncoder()
```

```python
y=ly.fit_transform(y)
```

```python
ly.classes_
```

```python
print("0 is Representing "+ ly.inverse_transform([0]))
print("1 is Representing " + ly.inverse_transform([1]))
print("2 is Representing " + ly.inverse_transform([2]))
print("3 is Representing " + ly.inverse_transform([3]))
print("4 is Representing " + ly.inverse_transform([4]))
print("5 is Representing " + ly.inverse_transform([5]))
```

```python
#Now we will split the data into training and testing data using the train_test_split function
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33,
                                                    random_state=7)
```

***Decsion Tree Algorithm***


```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
from sklearn import metrics
predictions = dt.predict(X_test)
print("Accuracy_Score of Testing Data =", format(metrics.accuracy_score(y_test, predictions)))
```

```python
#Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix")
print()
print(confusion_matrix(y_test, predictions))
```

```python
print("Classification Report")
print()
print(classification_report(y_test,predictions))
```

```python
#Checking our own and Algorithms Predictions
check = dt.predict(X_test)
print("Algorithm's Prediction on index 0")
print(check[0])
print("We Checking Values of Index 0")
print(y_test[0])
```
